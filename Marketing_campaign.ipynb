```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.get_dummies(data,columns=['Education', 'Marital_Status'])
```

```python

df.dtypes
```

```python
df.columns
```

```python
df1=df.drop(['Education_2n Cycle', 'Education_Basic', 'Education_Graduation',
       'Education_Master', 'Education_PhD', 'Marital_Status_Absurd',
       'Marital_Status_Alone', 'Marital_Status_Divorced',
       'Marital_Status_Married', 'Marital_Status_Single',
       'Marital_Status_Together', 'Marital_Status_Widow',
       'Marital_Status_YOLO'], axis= 1)
```

```python
df1.dtypes
```

```python
df1=df1.drop(['Dt_Customer'], axis=1)
```

```python
df1.head()
```

```python
data_cor= df1.corr()
data_cor
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(data_cor, annot=True, cmap="coolwarm")
plt.show()
```

```python
X = df1.drop(['Response'], axis=1)
Y = df1['Response']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=6)
knn_model.fit (X_train, Y_train)
```

```python
pred= knn_model.predict(X_test)
pred
```

```python
print('The accuracy of the KNN with K=6 is {}%'.format(round(accuracy_score(pred,Y_test)*100,2)))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, Y_train)
```

```python
pred1=model.predict(X_test)
pred1
```

```python
print('The accuracy of NB is {}%'.format(round(accuracy_score(pred1,Y_test)*100,2)))
```
